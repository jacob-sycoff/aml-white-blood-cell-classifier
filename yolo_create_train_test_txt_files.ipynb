```python
import os 
from collections import defaultdict
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#file paths
images_file_path = '/content/gdrive/MyDrive/yolo_data/AML_Cytomorphology/images' #these are the original images from TCIA. We copied orginal data made a yolo folder with data.
labels_file_path = '/content/gdrive/MyDrive/yolo_data/AML_Cytomorphology/labels' #deprecated path that was tested to contain labels for all images 
train_images_path = '/content/gdrive/MyDrive/haley_yolo_split_data/images/train' #path for yolo train images
test_images_path = '/content/gdrive/MyDrive/haley_yolo_split_data/images/test' #path for yolo test images
train_labels_path = '/content/gdrive/MyDrive/haley_yolo_split_data/labels/train' #path for yolo train txt files
test_labels_path = '/content/gdrive/MyDrive/haley_yolo_split_data/labels/test' #path for yolo test txt files
```

```python
#dictionary for translating classes to integers
classes_dictionary = {'BAS':0, 'EBO':1, 'EOS':2, 'KSC':3, 'LYA':4, 'LYT':5, 'MMZ':6, 'MOB':7, 'MON':8, 'MYB':9, 'MYO':10, 'NGB':11, 'NGS':12, 'PMB':13, 'PMO':14 }
```

## Create Txt Files for Training Images

```python
for file in os.listdir(train_images_path):
    filename = file[:-5] #file name without tiff extension so just cell type and number
    cell_type = filename[:3] #name of cell type 
    cell_class = classes_dictionary[cell_type] #number for cell class type
    #Write txt file into labels folder with appropriate class name
    with open(train_labels_path + "/" + filename + ".txt", "w") as f:
        f.write(str(cell_class) + " " "0.5 0.5 1.0 1.0") #0.5 and 0.5 for center x and y coordinates and 1.0 and 1.0 for full width and height of images - check later for correctness
```

```python
#check to see how many train txt files there are 
print(len([name for name in os.listdir(train_labels_path) if os.path.isfile(os.path.join(train_labels_path, name))]))
```

## Create Txt Files for Test Images

```python
for file in os.listdir(test_images_path):
    filename = file[:-5] #file name without tiff extension so just cell type and number 
    cell_type = filename[:3] #name of cell type 
    print(cell_type)
    cell_class = classes_dictionary[cell_type] #number for cell class type 
    #Write txt file into labels folder with appropriate class name
    with open(test_labels_path + "/" + filename + ".txt", "w") as f:
        f.write(str(cell_class) + " " "0.5 0.5 1.0 1.0") #0.5 and 0.5 for center x and y coordinates and 1.0 and 1.0 for full width and height of images - check later for correctness
```

```python
#check to see how many test txt files there are 
print(len([name for name in os.listdir(test_labels_path) if os.path.isfile(os.path.join(test_labels_path, name))]))
```

## Check Breakdown of Text Files for Train and Test Images by Cell Type

```python
#Look at breakdown of images in training set and test set by cell type 
train_dic = defaultdict(int)
test_dic = defaultdict(int)
for file in os.listdir(train_labels_path):
    train_dic[file[:3]] += 1 
for file in os.listdir(test_labels_path):
    test_dic[file[:3]] +=1 
```

```python
print(train_dic)
print(test_dic)
```

```python
data = {'Cell Type': train_dic.keys(),
        'Train': train_dic.values(),
      'Test': [test_dic[key] for key in train_dic.keys()]}
```

```python
df = pd.DataFrame(data)
df.sort_values('Train', inplace = True, ascending = False, ignore_index = True)
df
```

```python
df['Total'] = df['Train'] + df['Test']
df
```
